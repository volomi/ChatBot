```python
for story, question, answer in all_data:
    vocab = vocab.union(set(story))
    vocab = vocab.union(set(question))
```

```python
vocab.add('yes')
vocab.add('no')
```

```python
vocab_length = len(vocab)+1
```

```python
max_story_len = max([len(data[0]) for data in all_data])
```

```python
max_ques_len = max([len(data[1]) for data in all_data])
max_ques_len 
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer =Tokenizer(filters =[])
```

```python
tokenizer.fit_on_texts(vocab)
```

```python
tokenizer.word_index #roll-assign
```

```python
train_story_text = []
train_question_text = []
train_answers = []

for story, question, answer in train_data:
    train_story_text.append(story)
    train_question_text.append(question)
```

```python
train_story_seq = tokenizer.texts_to_sequences(train_story_text)
```

```python
len(train_story_text)
```

```python
train_story_seq
```

```python
#Functionalize the vectorization
def vectorize_stories(data, word_index = tokenizer.word_index,
                     max_story_len = max_story_len, max_ques_len = max_ques_len):

    X = [] #stories
    Xq = [] #query
    Y = [] #answer

    for story, query, answer in data:
        x = [word_index[word.lower()] for word in story]
        xq = [word_index[word.lower()] for word in story]
        y = np.zeros(len(word_index) +1)
        y[word_index[answer]] =1
    
        X.append(x)
        Xq.append(xq)
        Y.append(y)
    
    return(pad_sequences(X, maxlen = max_story_len),
           pad_sequences(Xq, maxlen = max_ques_len),
           np.array(Y))
```

```python
inputs_train, queries_train, answers_train = vectorize_stories(train_data)
```

```python
inputs_test, queries_test, answers_test = vectorize_stories(test_data)
```

```python
from keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding
from keras.layers import Input, Activation, Dense, Permute, Dropout, add, dot, concatenate, LSTM
```

```python
input_sequence = Input((max_story_len,))
question = Input((max_ques_len,))
```

```python
#Input Encoder m
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim = vocab_length, output_dim = 64))
input_encoder_m.add(Dropout(0.3))
```

```python
#Input Encoder c
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim = vocab_length, output_dim = max_ques_len))
input_encoder_c.add(Dropout(0.3))
```

```python
#Question Encoder
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim = vocab_length, output_dim = 64, input_length = max_ques_len))
question_encoder.add(Dropout(0.3))
```

```python
#Encode Sequence
input_encoded_m = input_encoder_m(input_sequence)
input_encoded_c = input_encoder_c(input_sequence)
question_encoded = question_encoder(question)
```

```python
match = dot([input_encoded_m,question_encoded], axes = (2,2))
match = Activation('softmax')(match)
```

```python
response = add([match,input_encoded_c])
response = Permute((2,1))(response)
```

```python
#Concatenate
answer = concatenate([response,question_encoded])
```

```python
answer
```

```python
answer = LSTM(32)(answer)
```

```python
#Regularize
answer = Dropout(0.5)(answer)
answer = Dense(vocab_length)(answer)
```

```python
answer = Activation('softmax')(answer)
```

```python
model = Model([input_sequence, question], answer)
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit([inputs_train, queries_train], answers_train,
                   batch_size = 32, epochs =120, 
                   validation_data = ([inputs_test, queries_test], answers_test)
                   )
```

```python
import matplotlib.pyplot as plt
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.xlabel("epochs")
plt.ylabel("Accuracy")
```
